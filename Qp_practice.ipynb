```python
#E-commerce Discount Calculator

amount=int(input("enter the purchase amount:"))
if amount>=100 and amount<500:
    dis=amount*0.1
    print(f"amount to be paid is {amount-dis}")
elif amount >= 500:
    dis=amount*0.2
    print(f"amount to be paid is {amount-dis}")
else:
    print("pay the full amount, no discount",amount)
```

```python
#Traffic Light Simulation

col=["red","yellow","green"]
colin=input("enter any colour between: /nred /nyellow /ngreen")
if colin==col[0]:
    print("STOP")
elif colin==col[1]:
    print("ready to move")
elif colin==col[2]:
    print("GO")
else:
    print("invalid colour")
```

```python
# Write a program that accepts the number of units consumed and calculates the total bill.

units=int(input("enter the units you used:"))
ba=0
if units>0 and units<=100:
    ba=units*0.5
elif units>100 and units<=200:
    diff=units-100
    ba=(diff*0.75)+(100*0.5)
elif units>200:
    diff=units-200
    ba=(100*0.5)+(100*0.75)+(diff*1)
else:
    print("invalid")
print("bill amount is ",ba)    
```

```python
# Write a program that takes the temperature as input and displays the corresponding alert message.

temperature = float(input("Enter the temperature in °C: "))
if temperature < 0:
    print("Freezing weather")
elif 0 <= temperature <= 20:
    print("Cold weather")
elif 21 <= temperature <= 30:
    print("Warm weather")
else:
    print("Hot weather")
```
