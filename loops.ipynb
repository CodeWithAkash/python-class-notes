```python
#checking palindrome

num=121
temp=num
rev=0
while(num>0):
    rem=num%10
    rev=(rev*10)+rem
    num=num//10
if rev==temp:
    print("palindrome")
else:
    print("not")
    
```

```python
#check a number is prime or not 

num = int(input("enter a number:"))
count=0
for i in range(2,num):
    if num % i==0:
        count+=1
    else:
        count=count
        
if count==0:
    print("prime")
else:
    print("not prime")
    
```
