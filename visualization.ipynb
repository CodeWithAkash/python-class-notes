```python
#matplotlib and seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
df
```

```python
plt.figure()
plt.hist(df['total_bill'])
plt.title("Histogram")
plt.xlabel("Total bill")
plt.ylabel("Frequency")
plt.show()
```

```python
df['total_bill'].max()
```

```python
df['total_bill'].min()
```

```python
df['total_bill'].mean()
```

```python
df['total_bill'].median()
```

```python
#KDE Plot:
sns.kdeplot(df['total_bill'], fill=True)
plt.title("KDE plot")
plt.xlabel("Total bill")
plt.ylabel("Frequency")
plt.show()    
```

```python
plt.figure()
sns.boxplot(x=df['total_bill'])
plt.title("Box Plot")
plt.xlabel("total bill")
plt.show()
```

```python
plt.figure()
sns.violinplot(x=df['total_bill'])
plt.title("Violin Plot")
plt.show()
```

```python
plt.figure()
sns.barplot(x="day",y="total_bill",data=df)
plt.title("Bar Plot")
plt.show()
```

```python
plt.figure()
sns.barplot(x="total_bill",y="day",data=df)
plt.title("Bar Plot")
plt.show()
```

```python
plt.figure()
df['day'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
sns.countplot(x='day',data=df)
plt.title('Count plot')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title('Scatter plot')
plt.show()
```

```python
plt.plot(df['total_bill'])

plt.title("line plot")
plt.show()
```

```python
plt.scatter(df['total_bill'],df['tip'],s=df['size']*30)
plt.title('bubble sort')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
corr.round(2)
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```
