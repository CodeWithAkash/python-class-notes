```python
#matplotlib and seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
df
```

```python
plt.figure()
plt.hist(df['total_bill'])
plt.title("Histogram")
plt.xlabel("Total bill")
plt.ylabel("Frequency")
plt.show()
```
