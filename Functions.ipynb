```python
#without return and input

def prime1():
    for num in range(2, 21):
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)   
prime1()
    
```

```python
#with return and without input
def prime2a():
    for i in range(2,21):
        for j in range(2,i):
            if i%j==0:
                break
        else:
            return i
val=prime2a()            
print(val)            
```

```python
#with return without input using list

def prime2b():
    num=[]
    for i in range(2,21):
        for j in range(2,i):
            if i%j==0:
                break
        else:
           num.append(i)
    return num
val=prime2b()        
print(val)
```

```python
def prime3(n):
    value=[]
    for j in range(2,n+1):
        for i in range(2,j):
            if j%i ==0:
                break
        else:
            value.append(j)
    return value
val=prime3(20)
print(val)
```

```python
#with input and with return 

def prime3(a,b):
    val=[]
    for j in range(2,b+1):
        for i in range(2,j):
            if j%i ==0:
                break
        else:
            val.append(j)
    return val        
prime3(2,20)
```

```python
#with input and without return

def prime4(a,b):
    for i in range(a,b+1):
        for j in range(2,i):
            if i%j ==0:
                break
        else:
            print(i)
prime4(2,20)            
```

```python
#a program to create fibonnacci series

def fibo():
    a=0
    b=1
    for i in range(1,9):
        print(a)
        s=a+b
        a,b=b,s
fibo()            
```

```python
#sum of n numbers using self calling 
def sum_n(n):
    if n == 0:
        return 0
    return n+sum_n(n-1)   
sum_n(10)    
```
