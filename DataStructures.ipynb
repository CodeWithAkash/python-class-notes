```python
#create a list of even num,odd num, prime num from 1 to 20

even=[]
odd=[]
prime=[]
for i in range(1,21):
    if i %2==0:
        even.append(i)
    else:
        odd.append(i)

    if i!=1:
        for j in range(2,i):
            if i%j ==0:
                break
        else:
            prime.append(i)
print(even,odd, prime)            
            
```
