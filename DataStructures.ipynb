```python
#create a list of even num,odd num, prime num from 1 to 20

even=[]
odd=[]
prime=[]
for i in range(1,21):
    if i %2==0:
        even.append(i)
    else:
        odd.append(i)

    if i!=1:
        for j in range(2,i):
            if i%j ==0:
                break
        else:
            prime.append(i)
print(even,odd, prime)            
            
```

```python
#count the number of char, num of digits, and num if special char
s1="python@123?best:64"
intt=0
charr=0
sp=0
for i in s1:
    if i.isnumeric():
        intt += 1
    elif i.isalpha():
        charr +=1
    else:
        sp+=1
print(intt,charr,sp)
```
