```python
#E-commerce Discount Calculator

amount=int(input("enter the purchase amount:"))
if amount>=100 and amount<500:
    dis=amount*0.1
    print(f"amount to be paid is {amount-dis}")
elif amount >= 500:
    dis=amount*0.2
    print(f"amount to be paid is {amount-dis}")
else:
    print("pay the full amount, no discount",amount)
```

```python
#Traffic Light Simulation

col=["red","yellow","green"]
colin=input("enter any colour between: /nred /nyellow /ngreen")
if colin==col[0]:
    print("STOP")
elif colin==col[1]:
    print("ready to move")
elif colin==col[2]:
    print("GO")
else:
    print("invalid colour")
```

```python
#Odd or Even and Divisibility Check

num=int(input("enter a number"))
if num%2==0:
    print("even")
else:
    print("odd")
if num%5==0:
    print("divisible by 5")
else:
    print("not divisble by 5 ")
```

```python
#Password Strength Checker

pas=input("enter your password")
if len(pas) >= 8:
    print("strong")
elif len(pas) >5 and len(pas) <7:
    print("medium")
else:
    print("weak")
    
```

```python
# Write a program that accepts the number of units consumed and calculates the total bill.

units=int(input("enter the units you used:"))
if units>0 and units<=100:
    print(f"you total bill is {units*0.5}")
elif units>100 and units<=200:
    print(f"you total bill is {units*0.75}")   
elif units>200:
    print(f"you total bill is {units*1}")    
else:
    print("invalid")
```

```python
#Write a program to check whether a person is eligible for the loan based on their age and monthly income.

age=int(input("enter your age"))
income=int(input("enter you income:(yearly)"))
if 21<=age<60 and income>=5000:
    print("verified age and income, thus eligible for loan process")
else:
    print("invalid")
```

```python
# Write a program that takes the temperature as input and displays the corresponding alert message.

temperature = float(input("Enter the temperature in °C: "))
if temperature < 0:
    print("Freezing weather")
elif 0 <= temperature <= 20:
    print("Cold weather")
elif 21 <= temperature <= 30:
    print("Warm weather")
else:
    print("Hot weather")
```
