```python
cit=input("enter your country ")
cit=cit.lower()
age=int(input("enter your age"))
if cit=="india" and age >= 18:
    print("you are eligible to vote")
elif cit=="india" and age<18:
    print("you are underaged")
else:
    print("not verified country")
    
```
