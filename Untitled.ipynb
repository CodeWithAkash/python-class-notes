```python
#create a list of even num,odd num, prime num from 1 to 20

even=[]
odd=[]
prime=[]
for i in range(1,21):
    if i %2==0:
        even.append(i)
    else:
        odd.append(i)

for j in range(2,21):
        for i in range(2,j):
            if j%i ==0:
                break
        else:
            prime.append(j)
print(even,odd, prime)            
            
```
