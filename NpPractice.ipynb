```python
data = np.random.randint(10, 101, 100)


mean = np.mean(data)
std = np.std(data)

# Step 3: calculate Z-scores
z_scores = (data - mean) / std

# Step 4: detect outliers (|z| > 3)
outliers = data[np.abs(z_scores) > 3]
```
