```python
data = np.random.randint(10, 101, 100)


mean = np.mean(data)
std = np.std(data)


z_scores = (data - mean) / std


outliers = data[np.abs(z_scores) > 3]
outliers
```

```python
#normalizing using min-max

a=np.array([120, 150, 200, 250, 300])
normalized = (a - np.min(a)) / (np.max(a) - np.min(a))
normalized
```
